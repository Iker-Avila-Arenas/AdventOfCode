```python
import re
```

```python
def remove_empty(l):
    if l[0]=="" and l[-1]=="":
        return l[1:-1]
    elif l[0]=="":
        return l[1:]
    elif l[-1]=="":
        return l[:-1]
    else:
        return l
```

```python
input = """123 328  51 64 
 45 64  387 23 
  6 98  215 314
*   +   *   +  """
input = input.split("\n") #n
input = input[::-1] #n
signs = re.split(r' +', input[0]) #m
signs = remove_empty(signs)
m = len(signs) #m
results = []
for i in range(m): #m
    if signs[i]=="+":
        results.append(0)
    else:
        results.append(1)
print(results)
for line in input[1:]: #n
    line_list = remove_empty(re.split(r' +', line))
    for i in range(m): #m
      if signs[i]=="+":
          results[i]+=int(line_list[i])
      else:
          results[i]*=int(line_list[i])
print(sum(results))
# Code runs in O(m*n)
```
