```python
def valid_id2(id):
    sid = str(id)
    length=len(sid)
    # decomposition of length of string (prime decomposition can make this faster)
    for n in range(1,length//2+1):
        if length%n==0:
            elems = length//n
            l = [sid[n*i:n*i+n] for i in range(elems)]
            if l.count(l[0])==elems:
                       return True
            
```

```python
def remove_empty(l):
    if l[0]=="" and l[-1]=="":
        return l[1:-1]
    elif l[0]=="":
        return l[1:]
    elif l[-1]=="":
        return l[:-1]
    else:
        return l
```

```python
input = """123 328  51 64 
 45 64  387 23 
  6 98  215 314
*   +   *   +  """
input = input.split("\n") #n
input = input[::-1] #n
signs = re.split(r' +', input[0]) #m
signs = remove_empty(signs)
m = len(signs) #m
results = []
for i in range(m): #m
    if signs[i]=="+":
        results.append(0)
    else:
        results.append(1)
print(results)
for line in input[1:]: #n
    line_list = remove_empty(re.split(r' +', line))
    for i in range(m): #m
      if signs[i]=="+":
          results[i]+=int(line_list[i])
      else:
          results[i]*=int(line_list[i])
print(sum(results))
# Code runs in O(m*n)
```

```python
def eudist(r1,r2):
    eudist2 = 0
    for i in range(3):
        eudist2 += (r2[i]-r1[i])**2
    return eudist2
```

```python
input = """162,817,812
57,618,57
906,360,560
592,479,940
352,342,300
466,668,158
542,29,236
431,825,988
739,650,466
52,470,668
216,146,977
819,987,18
117,168,530
805,96,715
346,949,466
970,615,88
941,993,340
862,61,35
984,92,344
425,690,689"""

with open("day8.txt") as f:
    input = f.read()

start_time = time.time()

input = input.split("\n") # m*n
temp = []
for j in input: # n
    temp.append([int(i) for i in j.split(",")])
input = temp

# find the m smallest values in one side of the matrix of distances (no need to compute the whole matrix)
m=1000
min_distances = [(float('inf'), float('inf'), float('inf'))]*m
for j in range(len(input)): # n 
    for i in range(j+1, len(input)): #n/2
        distance = eudist(input[j],input[i])
        # we assume the min_distances is always sorted from largest to smallest
        idx = 0
        while idx<m and min_distances[idx][2]>distance: # m
            idx+=1
        if idx!=0:
            min_distances = min_distances[1:idx]+[(i,j,distance, input[i], input[j])]+min_distances[idx:]
# previous code has complexity n^2*m/2 (/2 for half matrix)

# create clusters of connected box
connected = []
for j in range(len(min_distances)): # m
    k = min_distances[j][0]
    v = min_distances[j][1]
    k_c = None
    v_c = None
    for i in range(len(connected)): # if all connected pair to pair, results in 500 clusters, so the mean would be 250 = m/4
        if k in connected[i]:
            k_c = i
        if v in connected[i]:
            v_c = i
        if v_c != None and k_c != None:
            break
    # case 1, none of the values appears in connected --> new cluster
    if k_c==None and v_c==None:
        connected.append([k,v])
    # case 2 and 3, only one of them appears in connectd --> Add the other one to the existing cluster
    if k_c==None and v_c!=None:
        connected[v_c].append(k)
    elif k_c!=None and v_c==None:
        connected[k_c].append(v)
    # case 4, when both are already part of same cluster --> Do nothing
    elif k_c!=None and k_c==v_c:
        None
    # case 5, when both are already part of different clusters --> Combine clusters
    elif k_c!=None and v_c!=None and k_c!=v_c: # reddundant if for clarification
        if k_c>v_c:
            l1 = connected.pop(k_c)
            l2 = connected.pop(v_c)
        else:
            l1 = connected.pop(v_c)
            l2 = connected.pop(k_c)
        connected.append(l1+l2)
# previous code has complexity m^2

# find the three largest clusters:
lengths = []
for i in connected: # k = len(connected)=300
    s = len(i)
    lengths.append(s)
lengths.sort() # k*log(k)
end_time = time.time()
print(lengths[-3]*lengths[-2]*lengths[-1])
print(end_time-start_time)
```
