```python
pos = 50
movements =  ["L68", "L30", "R48", "L5", "R60", "L55", "L1", "L99", "R14", "L82"]
count = 0

for i in movements:
    if i[0]=="L":
        n = -int(i[1:])
    else:
        n = int(i[1:])
    pos = (pos+n)%100
    count+=1 if pos==0 else 0
    print(pos, count)
print(count)
```

```python
pos=50
count=0

with open("day1.txt") as file:
    while i:=file.readline()[:-1]:
        if i[0]=="L":
            n = -int(i[1:])
        else:
            n = int(i[1:])
        pos = (pos+n)%100
        count+=1 if pos==0 else 0

print(count)

#Only travel through the lines of the file one time, so complexity is Θ(n)
```

Other classical option:

```python
pos=50
count=0

with open("day1.txt") as file:
    list = file.read().splitlines()

for i in list:
    if i[0]=="L":
        n = -int(i[1:])
    else:
        n = int(i[1:])
    pos = (pos+n)%100
    count+=1 if pos==0 else 0
print(count)

# Have to go through the elements 2 times, one to extract the list and the other to do the calculation, so Θ(2n)
```

```python
def valid_id(id):
    sid = str(id)
    mid = len(sid)//2
    return sid[:mid]==sid[mid:]
```

```python
input = "11-22,95-115,998-1012,1188511880-1188511890,222220-222224,1698522-1698528,446443-446449,38593856-38593862,565653-565659,824824821-824824827,2121212118-2121212124"
ranges = input.split(",")
count=0
for r in ranges:
    ids = r.split("-")
    for i in range(int(ids[0]),int(ids[1])+1,1):
        count+=i if valid_id(i) else 0
print(count)
```

```python
def valid_id2(id):
    sid = str(id)
    length=len(sid)
    # decomposition of length of string (prime decomposition can make this faster)
    for n in range(1,length//2+1):
        if length%n==0:
            elems = length//n
            l = [sid[n*i:n*i+n] for i in range(elems)]
            if l.count(l[0])==elems:
                       return True
            
```

```python

input = """..@@.@@@@.
@@@.@.@.@@
@@@@@.@.@@
@.@@@@..@.
@@.@@@@.@@
.@@@@@@@.@
.@.@.@.@@@
@.@@@.@@@@
.@@@@@@@@.
@.@.@@@.@.
"""

grid = input.splitlines()
height = len(grid)
width = len(grid[0])
```

```python
def try_remove(a,b,grid):
    i_look = [a+n for n in range(-1,2) if (a+n)>=0 and (a+n)<height]
    j_look = [b+n for n in range(-1,2) if (b+n)>=0 and (b+n)<width]
    l = []
    for i in i_look:
        for j in j_look:
            l.append(grid[i][j])
    if l.count("@")<=4: #= beacuse I'm counting myself
        return True
    return False
```

```python
total_count = 0
while True:
    count = 0
    for a in range(height):
        for b in range(width):
            if grid[a][b]=="@" and try_remove(a,b,grid):
                lin_list = list(grid[a])
                lin_list[b] = "."
                grid[a]= "".join(lin_list)
                count+=1
    total_count+=count
    if count==0:
        break
print(total_count)

# Hiper ineficiente, pero funciona, hay que probar en input
```
