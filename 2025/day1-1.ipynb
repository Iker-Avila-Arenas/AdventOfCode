### Part 1

Test

```python
pos = 50
movements =  ["L68", "L30", "R48", "L5", "R60", "L55", "L1", "L99", "R14", "L82"]
count = 0

for i in movements:
    if i[0]=="L":
        n = -int(i[1:])
    else:
        n = int(i[1:])
    pos = (pos+n)%100
    count+=1 if pos==0 else 0
    print(pos, count)
print(count)
```

On input:

```python
pos=50
count=0

with open("day1.txt") as file:
    while i:=file.readline()[:-1]:
        if i[0]=="L":
            n = -int(i[1:])
        else:
            n = int(i[1:])
        pos = (pos+n)%100
        count+=1 if pos==0 else 0

print(count)

#Only travel through the lines of the file one time, so complexity is Θ(n)
```

Other classical option:

```python
pos=50
count=0

with open("day1.txt") as file:
    list = file.read().splitlines()

for i in list:
    if i[0]=="L":
        n = -int(i[1:])
    else:
        n = int(i[1:])
    pos = (pos+n)%100
    count+=1 if pos==0 else 0
print(count)

# Have to go through the elements 2 times, one to extract the list and the other to do the calculation, so Θ(2n)
```

```python
pos = 50
movements =  ["L68", "L30", "R48", "L5", "R60", "L55", "L1", "L99", "R14", "L82"]
count = 0

for i in movements:
    if i[0]=="L":
        n = -int(i[1:])
    else:
        n = int(i[1:])
    new_pos = (pos+n)%100
    count+= abs(pos+n)//100
    if pos*(pos+n)<0 or pos+n==0:
        count+=1
    print(pos, n, pos+n, (pos+n)%100, abs(pos+n)//100, count)
    pos=new_pos

print(count)
```

Input:
